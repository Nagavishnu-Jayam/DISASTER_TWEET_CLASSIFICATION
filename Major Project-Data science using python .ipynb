```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
###reading dataset
df=pd.read_csv('disaster_tweets_data.csv')
df.head()
```

```python
#removal of null values if present any
df.isnull().sum()  #since we are not having any null values we can neglet this section
```

```python
#removing patterns in input text
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for word in r:
        input_txt = re.sub(word, "", input_txt)
    return input_txt
```

```python
df['processed_data'] = np.vectorize(remove_pattern)(df['tweets'], "#[\w]*")
df.head()
```

```python
#removing punctuations
# now we will remove all the special characters, punctuations and numbers
df['processed_data'] = df['processed_data'].str.replace("[^a-zA-Z#]", " ")
df.head()
```

```python
# converting all the  words to lower case
df['processed_data'].str.lower()
```

```python
# Tokenizing words
tokens = df['processed_data'].apply(lambda x: x.split())
tokens.head()
```

```python
#Removing Stop words
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")

sw = stopwords.words('English')

re_sw = [i for i in df['processed_data'] if i not in sw]
print(re_sw)
```

```python
#Stemming or lemmatizing the words

from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer
ps = PorterStemmer() 
lm= WordNetLemmatizer()

#Stemming
stem_ps = [ps.stem(i) for i in df['processed_data']]
print(stem_ps)
```

```python
#lemmatizing
nltk.download('wordnet')
nltk.download('omw-1.4')
lem = [lm.lemmatize(i) for i in df['processed_data']]
print(lem)
```
